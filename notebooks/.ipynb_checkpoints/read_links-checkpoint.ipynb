```python
url = 'https://www.naehrwertrechner.de/naehrwerte/X548241/Zwiebeln+ged%C3%BCnstet%2C+in+saurer+Sahne+%28Zubereitung+Haushalt%29'
```

```python
chrome_driver.get(url)
page_content = BeautifulSoup(chrome_driver.page_source, "lxml")
page_items = page_content.find_all('tbody')
```

```python
item_properties = {'item_descr': page_content.find_all('h2')[0].text,
                  'item_title': page_content.find('title').text}

for page_item in page_items:
    for properties in page_item.find_all('tr'):
        data = [p.text.strip() for p in properties.find_all('td')]
        property_name = data[0]
        property_value = data[1]
        
        if property_name is not None and property_value is not None:
            if property_name.startswith('Tagesbedarf'):
                continue
                
            if property_name not in accepted_property_names:
                print("Skipping: %s" % (properties))
                continue
                
            if property_name == 'Broteinheiten' or property_name == 'Carbohydrate units':
                country_src = properties.find('img').get('src')
                country_code = country_src.replace('.png', '')[-2:].upper()
                gram = property_value.endswith('mg')
                suffix = '_'
                if country_code is not None:
                    suffix += country_code
                if gram:
                    suffix += '_MG'
                else:
                    suffix += '_UNIT'
                property_name += suffix
                accepted_property_names.append(property_name)
            
            item_properties[property_name] = property_value
```

```python
link_file = '../data/Bundeslms/nwr/V2/nwr.html'
with open(link_file, "r") as f:
    contents = f.read()
    soup = BeautifulSoup(contents, 'lxml')
```

```python
divs = soup.find_all('div', {'class': "11u"})
items = {}
for div in divs:
    link = div.find("a", href = True)
    link_name = link.text
    link_href = link['href']
    items[link_name] = link_href
```

```python
txt = '\n'.join(['%s\t%s\t%s' % (basename(dirname(value)), value, key) for (key, value) in items.items()])
output_file = '../data/BUndeslms/nwr/V2/nwr_links.txt'
with open(output_file, "w") as f:
    f.write(txt + '\n')
```
